# Matplotlib

Matplotlib 是一個**專業**的資料視覺化函式庫，滿地都是範例

```python
# 第一步就是引用專業的函示庫，對吧？
import matplotlib.pyplot as plt
```

```python
# 折線圖 Line，起手式
a = [1,2,6,5,2,1,8,5]
b = [2,4,6,2,6,2,3,5]

plt.plot(a)
plt.plot(b)
plt.show()
```

```python
# 散佈圖 Scatter，起手式

a = [[1,2], [3,2], [1,1], [3,1], [6,8]]

for data in a:
    plt.scatter(x=data[0], y=data[1])
plt.show()
```

```python
# 直方圖 Hist，起手式
a = [1,1,1,1,2,2,2,5,5,5,5,9,9,9,3,3,3]
plt.hist(a)
plt.show()
```

```python
# 直方圖 Bar，起手式
a = {"Apple": 20, "Banana": 200, "Peach": 50, "Monkey": 90}
plt.bar(a.keys(), a.values())
plt.show()
```

```python
words =['year,Tumor,Heart disease', '1986,118.3,87.4', '1987,118.9,94.3', '1988,121.1,87.3', '1989,121.7,82.2', '1990,115.3,83.9', '1991,117.8,84.1', '1992,121.9,86.4', '1993,125.5,79.8', '1994,126.9,72.6', '1995,136.4,64.7', '1996,143.5,62.2', '1997,144.3,56.8', '1998,140.5,55.4', '1999,138.3,54.5', '2000,141.6,48.8', '2001,143.1,48.8', '2002,144.2,48.5', '2003,143.1,47.9', '2004,142.8,50.1', '2005,141.2,48.3', '2006,139.3,43.8', '2007,142.6,44.4', '2008,133.7,51.7', '2009,132.5,47.7', '2010,131.6,47.4', '2011,132.2,47.9', '2012,131.3,47.9', '2013,130.4,47.7']
print(words)

death = []
for w in words:
    if w == 'year,Tumor,Heart disease':
        continue
    tmp = w.split(',') #這個學好喔!!
    tmp[1] = float(tmp[1])
    tmp[2] = float(tmp[2])
    death.append(tmp)
```

## Numpy

Python 中，專門處理矩陣運算的函式庫

```python
# 第一步就是引用專業的函示庫，對吧？
import numpy as np
```

```python
# 看起來就跟 list 一樣對吧？
na = np.array([1,2,3])
print(na)
```

```python
# 但 List 不能這樣做

na = na - 3
print(na)

na = na * 5
print(na)

na = na / 2
print(na)

na = na * na
print(na)
```

```python
# 常見用法：以成績輸入為例

score_name = np.array(["Alan", "Bob", "Cindy"])
score = np.array([10, 20, 30])

# 尋找分數小於 30 分的 "分數"
score_low = score[np.where(score < 30)]
print(score_low)
# 尋找分數小於 30 分的 "分數" 所對應的人
score_name_low = score_name[np.where(score < 30)]
print(score_name_low)
```

以下來結合 Matplotlib 的內容，順便來練習吧

```python
words =['year,Tumor,Heart disease', '1986,118.3,87.4', '1987,118.9,94.3', '1988,121.1,87.3', '1989,121.7,82.2', '1990,115.3,83.9', '1991,117.8,84.1', '1992,121.9,86.4', '1993,125.5,79.8', '1994,126.9,72.6', '1995,136.4,64.7', '1996,143.5,62.2', '1997,144.3,56.8', '1998,140.5,55.4', '1999,138.3,54.5', '2000,141.6,48.8', '2001,143.1,48.8', '2002,144.2,48.5', '2003,143.1,47.9', '2004,142.8,50.1', '2005,141.2,48.3', '2006,139.3,43.8', '2007,142.6,44.4', '2008,133.7,51.7', '2009,132.5,47.7', '2010,131.6,47.4', '2011,132.2,47.9', '2012,131.3,47.9', '2013,130.4,47.7']

death_year = np.array([])
death_tumor = np.array([])
death_heart = np.array([])
for w in words:
    if w == 'year,Tumor,Heart disease':
        continue
    tmp = w.split(',') #這個學好喔!!
    death_year = np.append(death_year,int(tmp[0])) #這裡我們把年份轉為 int
    death_tumor = np.append(death_tumor,float(tmp[1]))
    death_heart = np.append(death_heart,float(tmp[2]))

print(death_year)
print(death_tumor)
print(death_heart)
```

```python
plt.plot(death_tumor)
plt.plot(death_heart)
plt.xticks(range(death_year.shape[0]), death_year.astype(np.int), rotation=80)
plt.show()
```

```python
# Show 出 2000 年後的數據

plt.plot(death_tumor[np.where(death_year>=2000)])
plt.plot(death_heart[np.where(death_year>=2000)])
plt.xticks(range(death_year[np.where(death_year>=2000)].shape[0]), death_year[np.where(death_year>=2000)].astype(np.int), rotation=80)
plt.show()
```

## Exercise 1: Read Total profit of all months and show it using a line plot
<br>Total profit data provided for each month. Generated line plot must include the following properties: 
<br>X label name = Month Number
<br>Y label name = Total profit

```python
# 解答
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise Question 2: Read all product sales data and show it  using a multiline plot

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("sales_data.csv")
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.plot(monthList, faceCremSalesData,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(monthList, faceWashSalesData,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(monthList, toothPasteSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, bathingsoapSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, shampooSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, moisturizerSalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

# Exercise Question 3: Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("sales_data.csv")
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```
